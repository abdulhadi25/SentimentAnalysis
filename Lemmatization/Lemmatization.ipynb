## Lematization:

```python
# Reason:
# Better way than Stemming to obtain the original form of any given text rather than stemming 
# because lemmatization returns the actual word that has some meaning in the dictionary.
```

```python
#nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
lematzr = WordNetLemmatizer()
tweets_df['After Lematization'] = tweets_df['Removed Handlers'].apply(lambda x: ' '.join([lematzr.lemmatize(word,'v') for word in x.split() ]))
```
